```python
# we are only keeping columns that can become numeric
ndf = df.drop(
            ['year',
             #'quarter',
             #'month',
             #'dayofmonth',
             #'dayofweek',
             'uniquecarrier',
             #'airlineid',
             'tailnum',
             #'flightnum',
             #'originairportid',
             #'origincitymarketid',
             'origin',
             #'originstatefips',
             'originstatename',
             #'destairportid',
             #'destcitymarketid',
             'dest',
             #'deststatefips',
             'deststatename',
             #'crsdeptime',
             #'deptime',
             #'depdelay',
             #'depdelayminutes',
             #'depdel15',
             #'departuredelaygroups',
             #'crsarrtime',
             #'arrtime',
             #'arrdelay',
             #'arrdelayminutes',
             #'arrdel15',
             #'arrivaldelaygroups',
             #'cancelled',
             #'airtime',
             #'flights',
             #'distance',
             #'distancegroup',
             #'carrierdelay',
             #'weatherdelay',
             #'nasdelay',
             #'securitydelay',
             #'lateaircraftdelay',
             'tailnum',
             'manufacturer',
             #'mfr_year',
             'model']
             #'serial_number']
            ,axis = 1 )
```

```python
ndf['quarter'] = ndf['quarter'].astype(int)
ndf['month'] = ndf['month'].astype(int)
ndf['dayofmonth'] = ndf['dayofmonth'].astype(int)
ndf['dayofweek'] = ndf['dayofweek'].astype(int)
ndf['airlineid'] = ndf['airlineid'].astype(int)
ndf['flightnum'] = ndf['flightnum'].astype(int)
ndf['originairportid'] = ndf['originairportid'].astype(int)
ndf['origincitymarketid'] = ndf['origincitymarketid'].astype(int)
ndf['originstatefips'] = ndf['originstatefips'].astype(int) 
ndf['destairportid'] =  ndf['destairportid'].astype(int) 
ndf['destcitymarketid'] = ndf['destcitymarketid'].astype(int) 
ndf['deststatefips'] = ndf['deststatefips'].astype(int) 
ndf['crsdeptime'] = ndf['crsdeptime'].astype(int) 
ndf['deptime'] = pd.to_numeric(ndf['deptime'], errors='coerce').fillna(0,downcast='infer')
ndf['depdelay'] = pd.to_numeric(ndf['depdelay'], errors='coerce').fillna(0,downcast='infer')
ndf['depdelayminutes'] = pd.to_numeric(ndf['depdelayminutes'], errors='coerce').fillna(0)
ndf['depdel15'] = pd.to_numeric(ndf['depdel15'], errors='coerce').fillna(0)
ndf['departuredelaygroups'] = pd.to_numeric(ndf['departuredelaygroups'], errors='coerce').fillna(0)
ndf['crsarrtime'] = ndf['crsarrtime'].astype(int)
ndf['arrtime'] = pd.to_numeric(ndf['arrtime'], errors='coerce').fillna(0)
ndf['arrdelay'] = pd.to_numeric(ndf['arrdelay'], errors='coerce').fillna(0)
ndf['arrdelayminutes'] = pd.to_numeric(ndf['arrdelayminutes'], errors='coerce').fillna(0)
ndf['arrdel15'] = pd.to_numeric(ndf['arrdel15'], errors='coerce').fillna(0)
ndf['arrivaldelaygroups'] = pd.to_numeric(ndf['arrivaldelaygroups'], errors='coerce').fillna(0)
ndf['cancelled'] = pd.to_numeric(ndf['cancelled'], errors='coerce').fillna(0)
ndf['airtime'] = pd.to_numeric(ndf['airtime'], errors='coerce').fillna(0)
ndf['flights'] = pd.to_numeric(ndf['flights'], errors='coerce').fillna(0)
ndf['distance'] = pd.to_numeric(ndf['distance'], errors='coerce').fillna(0)
ndf['distancegroup'] = ndf['distancegroup'].astype(int)
ndf['carrierdelay'] = pd.to_numeric(ndf['carrierdelay'], errors='coerce').fillna(0,downcast='infer')
ndf['weatherdelay'] = pd.to_numeric(ndf['weatherdelay'], errors='coerce').fillna(0,downcast='infer')
ndf['nasdelay'] = pd.to_numeric(ndf['nasdelay'], errors='coerce').fillna(0,downcast='infer')
ndf['securitydelay'] = pd.to_numeric(ndf['securitydelay'], errors='coerce').fillna(0,downcast='infer')
ndf['lateaircraftdelay'] = pd.to_numeric(ndf['lateaircraftdelay'], errors='coerce').fillna(0,downcast='infer')
ndf['mfr_year'] = pd.to_numeric(ndf['mfr_year'], errors='coerce').fillna(0,downcast='infer')
ndf['serial_number'] = pd.to_numeric(ndf['serial_number'], errors='coerce').fillna(0,downcast='infer')
```
